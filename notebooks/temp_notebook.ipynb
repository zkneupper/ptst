```python
import ptst
```

```python
st = ptst.StyleTransferer()
```

### Load images

```python
lion_img_path = './data/hand_processed/images/lion_matthew-kerslake1.jpg'
space_img_path = './data/hand_processed/images/space_jeremy-thomas3.jpg'
```

```python
imsize = (128 * 2)
imsize
```

```python
st.set_content_img(lion_img_path, imsize=imsize)
st.set_style_img(space_img_path, imsize=imsize)
```

```python
assert st.check_image_size_match()
```

```python
# st.set_input_img()
st.input_img = st.style_img.clone()
```

```python
ptst.imshow(st.style_img)
ptst.imshow(st.content_img)
ptst.imshow(st.input_img)
```

### Model 

```python
st.initialize_cnn()
```

```python
st.execute_style_transfer(
    num_steps=5,
    style_weight=1000000,
    content_weight=1,
    file_path_output='./data/output/test_output_path/',
#     verbose=0
)
```
